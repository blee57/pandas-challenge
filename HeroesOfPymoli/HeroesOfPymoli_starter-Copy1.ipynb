```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
only_males = purchase_data.loc[purchase_data["Gender"] == "Male", :]
only_males_count = len(only_males["SN"].unique())
only_males_count

only_females = purchase_data.loc[purchase_data["Gender"] == "Female", :]
only_females_count = len(only_females["SN"].unique())

only_others = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]
only_others_count = len(only_others["SN"].unique())

totalcount = (only_males_count) + (only_females_count) + (only_others_count)

malespercentage = "{0: .2%}".format(int(only_males_count) / int(totalcount))
femalespercentage = "{0: .2%}".format(int(only_females_count) / int(totalcount))
otherspercentage = "{0: .2%}".format(int(only_others_count) / int(totalcount))

malefemaledicts = [{"": "Male","Total Count":(only_males_count),"Percentage of Players":(malespercentage)},
                  {"": "Female","Total Count":(only_females_count),"Percentage of Players":(femalespercentage)},
                  {"": "Other/Non-Disclosed","Total Count":(only_others_count),"Percentage of Players":(otherspercentage)}
                  ]

gender_df = pd.DataFrame(malefemaledicts)
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bin = [0 , 9 , 14, 19, 24, 29, 34, 39, 150]

age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bin,labels=age_labels)

age_group = purchase_data.groupby("Age Ranges")

age_group_purchase_df = age_group.count()
age_group_av_p_df = age_group.mean()
age_group_t_p_df = age_group.sum()

deletedcolumns11 = ['SN', 'Age', 'Gender', 'Item ID', 'Item Name', 'Price']
age_group_purchase_df_cleanup1 = age_group_purchase_df.drop(deletedcolumns11, 1)
age_group_purchase_df_cleanup2 = age_group_purchase_df_cleanup1.rename(columns={'Purchase ID':'Purchase Count'})

deletedcolumns12 = ['Age', 'Item ID', 'Purchase ID']
age_group_av_p_df_cleanup1 = age_group_av_p_df.drop(deletedcolumns12, 1)
age_group_av_p_df_cleanup2 = age_group_av_p_df_cleanup1.rename(columns={'Price':'Average Purchase Price'})

deletedcolumns13 = ['Age', 'Item ID', 'Purchase ID']
age_group_t_p_df_cleanup1 = age_group_t_p_df.drop(deletedcolumns12,1)
age_group_t_p_df_cleanup2 = age_group_t_p_df_cleanup1.rename(columns={'Price':'Total Purchase Value'})

merge_age_group_df = pd.merge(age_group_purchase_df_cleanup2, age_group_av_p_df_cleanup2, on="Age Ranges")
merge_age_group_df2 = pd.merge(merge_age_group_df,age_group_t_p_df_cleanup2, on="Age Ranges")
merge_age_group_df2

Agegroup1_1 = "${:,.2f}".format(int(age_group_t_p_df_cleanup2.iloc[0,0]) / int(age_group_df_cleanup2.iloc[0,0]))
Agegroup2_2 = "${:,.2f}".format(int(age_group_t_p_df_cleanup2.iloc[1,0]) / int(age_group_df_cleanup2.iloc[1,0]))
Agegroup3_3 = "${:,.2f}".format(int(age_group_t_p_df_cleanup2.iloc[2,0]) / int(age_group_df_cleanup2.iloc[2,0]))
Agegroup4_4 = "${:,.2f}".format(int(age_group_t_p_df_cleanup2.iloc[3,0]) / int(age_group_df_cleanup2.iloc[3,0]))
Agegroup5_5 = "${:,.2f}".format(int(age_group_t_p_df_cleanup2.iloc[4,0]) / int(age_group_df_cleanup2.iloc[4,0]))
Agegroup6_6 = "${:,.2f}".format(int(age_group_t_p_df_cleanup2.iloc[5,0]) / int(age_group_df_cleanup2.iloc[5,0]))
Agegroup7_7 = "${:,.2f}".format(int(age_group_t_p_df_cleanup2.iloc[6,0]) / int(age_group_df_cleanup2.iloc[6,0]))
Agegroup8_9 = "${:,.2f}".format(int(age_group_t_p_df_cleanup2.iloc[7,0]) / int(age_group_df_cleanup2.iloc[7,0]))

merge_age_group_df2['Avg Total Purchase per Person'] = [Agegroup1_1,Agegroup2_2,Agegroup3_3,Agegroup4_4
                                                        ,Agegroup5_5,Agegroup6_6,Agegroup7_7,Agegroup7_7]

merge_age_group_df2["Average Purchase Price"] = merge_age_group_df2["Average Purchase Price"].astype(float).map("${:,.2f}".format)
merge_age_group_df2["Total Purchase Value"] = merge_age_group_df2["Total Purchase Value"].astype(float).map("${:,.2f}".format)

merge_age_group_df2
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
deletedcolumns14 = ['Age','Gender','Item ID','Item Name']
purchase_data_d14_df = purchase_data.drop(deletedcolumns14, 1)

spender_df1 = purchase_data_d14_df.groupby("SN")

spender_df1_purchase = spender_df1.count()
spender_df2_average = spender_df1.mean()
spender_df3_tt = spender_df1.sum()

spender_df1_purchase_cleanup1 = spender_df1_purchase.drop('Price',1)
spender_df1_purchase_cleanup2 = spender_df1_purchase_cleanup1.rename(columns={'Purchase ID':'Purchase Count'})

spender_df2_average_cleanup1 = spender_df2_average.drop('Purchase ID',1)
spender_df2_average_cleanup2 = spender_df2_average_cleanup1.rename(columns={'Price':'Average Purchase Price'})

spender_df3_tt_cleanup1 = spender_df3_tt.drop('Purchase ID',1)
spender_df3_tt_cleanup2 = spender_df3_tt_cleanup1.rename(columns={'Price':'Total Purchase Value'})

merged_spender_df = pd.merge(spender_df1_purchase_cleanup2, spender_df2_average_cleanup2, on='SN')
merged_spender_df_1 = pd.merge(merged_spender_df, spender_df3_tt_cleanup2, on='SN')

merged_spender_df_sorted = merged_spender_df_1.sort_values("Total Purchase Value", ascending=False)

merged_spender_df_sorted['Average Purchase Price'] = merged_spender_df_sorted['Average Purchase Price'].astype(float).map("${:,.2f}".format)
merged_spender_df_sorted['Total Purchase Value'] = merged_spender_df_sorted['Total Purchase Value'].astype(float).map("${:,.2f}".format)

merged_spender_df_sorted.head()
```
